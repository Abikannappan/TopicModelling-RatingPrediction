## TOPIC MODELING

### LATENT DIRICHLET ALLOCATION (LDA)

```python
def sent_to_words(sentences):
    for sent in sentences:
        sent = re.findall('[A-Za-z]+',sent)  # Choose only words and numbers
        sent = simple_preprocess(str(sent), deacc=True) 
        yield(sent)  

# Convert to list
data = df_review.reviewText.values.tolist()
data_words = list(sent_to_words(data))
print(data_words[:1])
```

```python
stop_words = stopwords.words('english')
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
# Build the bigram and trigram models
bigram = gensim.models.Phrases(data_words, min_count=5, threshold=100) # higher threshold fewer phrases.
#trigram = gensim.models.Phrases(bigram[data_words], threshold=100)  
bigram_mod = gensim.models.phrases.Phraser(bigram)
#trigram_mod = gensim.models.phrases.Phraser(trigram)


# !python3 -m spacy download en  # run in terminal once
def process_words(texts, stop_words=stop_words, allowed_postags=['NOUN', 'ADJ', 'VERB', 'ADV']):
    """Remove Stopwords, Form Bigrams, Trigrams and Lemmatization"""
    texts = [w for w in texts if not w in stop_words]
    texts = [simple_preprocess(str(doc)) for doc in texts]
    texts = [bigram_mod[doc] for doc in texts]
    #texts = [trigram_mod[bigram_mod[doc]] for doc in texts]
    texts_out = []
    nlp = spacy.load('en_core_web_sm', disable=['parser', 'ner'])
    for sent in texts:
        doc = nlp(" ".join(sent)) 
        texts_out.append([token.lemma_ for token in doc if token.pos_ in allowed_postags])
    # remove stopwords once more after lemmatization
    texts_out=[[word for word in simple_preprocess(ps.stem(str(doc))) if word not in stop_words] for doc in texts_out]    
    return texts_out

data_ready = process_words(data_words)  # processed Text Data!
```

```python
# Create Dictionary
id2word = corpora.Dictionary(data_ready)

# Create Corpus: Term Document Frequency
corpus = [id2word.doc2bow(text) for text in data_ready]

# Build LDA model
```

```python
coherence_score={}
def ldamodel(num_topics):
    model = gensim.models.ldamodel.LdaModel(corpus=corpus,
                                               id2word=id2word,
                                               num_topics=num_topics, 
                                               random_state=100,
                                               passes=10,
                                               iterations=300,
                                               per_word_topics=True)
    # alpha and eta set to 1 to get better topics
    coherence_model_lda = CoherenceModel(model=model, texts=data_ready,  coherence='c_v')
    coherence_lda = coherence_model_lda.get_coherence()
    coherence_score[num_topics] = coherence_lda
    print('Number of Topics:',num_topics)
    #print(model.print_topics())
    return model
```

```python
def format_topics_sentences(ldamodel=None, corpus=corpus, texts=data):
    # Init output
    sent_topics_df = pd.DataFrame()
    # Get main topic in each document
    for i, row_list in enumerate(ldamodel[corpus]):
        row = row_list[0] if ldamodel.per_word_topics else row_list
        #print(row)
        row = sorted(row, key=lambda x: (x[1]), reverse=True)
        # Get the Dominant topic, Perc Contribution and Keywords for each document
        for j, (topic_num, prop_topic) in enumerate(row):
            if j == 0:  # => dominant topic
                wp = ldamodel.show_topic(topic_num)
                topic_keywords = ", ".join([word for word, prop in wp])
                sent_topics_df = sent_topics_df.append(pd.Series([int(topic_num), round(prop_topic,5), topic_keywords]), ignore_index=True)
            else:
                break
    sent_topics_df.columns = ['Dominant_Topic', 'Perc_Contribution', 'Topic_Keywords']

    # Add original text to the end of the output
    contents = pd.Series(texts)
    sent_topics_df = pd.concat([sent_topics_df, contents], axis=1)
    return(sent_topics_df)


```

```python
df_topic_keywords = format_topics_sentences(ldamodel=ldamodel(max_coherance_score), corpus=corpus, texts=data_ready)

# Format
df_dominant_topic = df_topic_keywords.reset_index()
df_dominant_topic.columns = ['Document_No', 'Dominant_Topic', 'Topic_Perc_Contrib', 'Keywords', 'Text']
df_dominant_topic.head(10)
```

***Above dataframe shows dominant topics in each document***

```python

topics = ldamodel(max_coherance_score).show_topics(formatted=False)

data_flat = [w for w_list in data_ready for w in w_list]
counter = Counter(data_flat)

out = []
for i, topic in topics:
    for word, weight in topic:
        out.append([word, i , weight, counter[word]])

df = pd.DataFrame(out, columns=['word', 'topic_id', 'importance', 'word_count'])        

# Plot Word Count and Weights of Topic Keywords
fig, axes = plt.subplots(max_coherance_score, 1, figsize=(50,70), sharey=True, dpi=160)
cols = [color for name, color in mcolors.TABLEAU_COLORS.items()]
for i, ax in enumerate(axes.flatten()):
    ax.bar(x='word', height="word_count", data=df.loc[df.topic_id==i, :], color=cols[i], width=0.2, alpha=0.3, label='Word Count')
    ax_twin = ax.twinx()
    ax_twin.bar(x='word', height="importance", data=df.loc[df.topic_id==i, :], color=cols[i], width=0.1, label='Weights')
    ax.set_ylabel('Word Count', color=cols[i],fontdict={'fontsize': 30,'fontweight': 30})
    ax_twin.set_ylim(0, 0.030); ax.set_ylim(0, 7000, 500)
    ax.set_title('Topic: ' + str(i), color=cols[i], fontsize=50)
    ax.tick_params(axis='y', left=False)
    ax.set_xticklabels(df.loc[df.topic_id==i, 'word'],fontdict={'fontsize': 30,'fontweight': 50},rotation=30, horizontalalignment= 'right')
    ax.legend(loc='upper left'); ax_twin.legend(loc='upper right')

fig.tight_layout(w_pad=2)    
fig.suptitle('Word Count and Importance of Topic Keywords', fontsize=60, y=1.05)    
plt.show()
```

***- Above plot shows weight and frequency of top words under each topics***

***- The plot also shows topic 0 as dominant plot***
